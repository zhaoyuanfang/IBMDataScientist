```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
import requests
import json
response = requests.post(url)
txt = response.text
soup = BeautifulSoup(txt)
```

```python
l = soup.find_all('table')
table = l[0]
```

```python
# define the dataframe columns
column_names = ['PostalCode', 'Borough'] 

# instantiate the dataframe
wiki = pd.DataFrame(columns=column_names)
```

```python
rows = table.find_all('tr')
for row in rows:
    row_ = row.find_all('td')
    if row_ and row_[1].string!='Not assigned' and row_[2].find('a'):
        wiki_row = {'PostalCode':row_[0].string, 'Borough':row_[1].string}
        wiki = wiki.append(wiki_row, ignore_index=True)
wiki = wiki.drop_duplicates()
wiki['Neighborhood']=''
wiki
```

```python
for row in rows:
    row_ = row.find_all('td')
    if row_ and row_[1].string!='Not assigned' and row_[2].find('a'):
#         wiki_row = {'PostalCode':row_[0].string, 'Borough':row_[1].string, 'Neighborhood':row_[2].find('a')['title']}
        index_l = wiki[(wiki.PostalCode == row_[0].string)&(wiki.Borough == row_[1].string)].index.tolist()
        index = index_l[0]
        wiki.iloc[index].Neighborhood += row_[2].find('a')['title']
```
